```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =sns.load_dataset('tips') ## this is a data set of tips 
```

```python
data
```

```python
data.head()
```

```python
data['total_bill']
```

```python
plt.plot(data['total_bill'])
plt.show()
```

```python
sns.lineplot(data['total_bill']) ## this is a line plot
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = data )  # this is a scatter plot and y is the total bill and x is the tip and data is the data set  
```

```python
sns.histplot(data['total_bill']) ## this is a histogram plot 
```

```python
plt.hist(data['total_bill']) #  this is a histogram plot by using matplotlib
```

```python
sns.kdeplot(data['total_bill']) # this is a kernel density plot 
```

```python
# catplot 
# relplot
```

```python
sns.relplot(x = "total_bill", y = "tip", data = data )  # this is a scatter plot and y is the total bill and x is the tip and data is the data set 
```

```python
data['time'].unique()
```

```python
sns.relplot(x = 'total_bill', y = "tip", hue = "time", data = data )  # this is a scatter plot and y is the total bill and x is the tip and data is the data set and hue is the time
```

```python
sns.relplot(x = 'total_bill', y = "tip", hue = "time", style = "smoker", data = data )  # this is a scatter plot and y is the total bill and x is the tip and data is the data set and hue is the time and style is the smoker
```

```python
sns.relplot(x = 'total_bill', y = "tip", hue = "time", style = "smoker",  kind = 'scatter', row = 'time',data = data )  # this is a scatter plot and y is the total bill and x is the tip and data is the data set and hue is the time and style is the smoker
```

```python
sns.relplot(y = 'total_bill', x = 'tip', hue = "time", style = "smoker", kind = 'scatter', col = 'smoker',data = data )  # this is a scatter plot and y is the total bill and x is the tip and data is the data set and hue is the time and style is the smoker
```

```python
data['day'].unique()
```

```python
sns.relplot(y = "total_bill", x = "tip", hue = 'time', data = data ,kind = "scatter", col = 'day')  # this is a scatter plot and y is the total bill and x is the tip and data is the data set and hue is the time and col is the day 
```

```python
sns.relplot(y = 'total_bill', x = 'tip', data = data )
plt.title('This is a scatter plot')
plt.xlabel('Total bill', fontsize = 20, color= 'black')
plt.ylabel('Tip', fontsize = 20, color= 'green')
plt.show()
```
