```python
img_arr_lst = []
labels = []
for key ,val in path_dict.items():
    for path in val:
        img_arr=cv2.imread(path)
        resized_img_arr=cv2.resize(img_arr,(224, 244))
        img_arr_lst.append(resized_img_arr)
        labels.append(key)  
    
```

```python
labels=np.array(labels)
```

```python
img_arr_lst=np.array(img_arr_lst)
```

```python
labels
```

```python
labels=np.array([1 if item =='dogs'else 0 for item in list(labels)])
```

```python
labels
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(img_arr_lst,labels,test_size=0.2)
```

```python
x_test
```

```python
x_train
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train.shape, y_train.shape
```

```python
# from tensorflow.keras.models import Sequential
# from tensorflow.layers import Conv2D,MaxPooling2D,Flatten,Dense

model =Sequential([
    # convolition layer 
    Conv2D(32,(3,3),activation='relu',input_shape=(244,224,3)),
    MaxPooling2D(2,2),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),



    Dense(64,activation='relu'),
    Dense(1,activation='sigmoid')


])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=10,batch_size=32)  
```

```python
model.summary()
```

```python
pred =model.predict(x_test)
```
